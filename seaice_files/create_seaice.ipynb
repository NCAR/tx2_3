# Create CICE grid
Use mom6_bathy to create CICE grid

```python
from mom6_bathy.grid import Grid
from mom6_bathy.topo import Topo
import xarray as xr
import datetime
import warnings
warnings.filterwarnings("ignore")
```

```python
today = datetime.date.today().strftime("%y%m%d")
print(today)
```

```python
grid = Grid.from_supergrid('../supergrid/ORCA_gridgen/ocean_hgrid_250930.nc')
```

```python
topo = Topo.from_topo_file(grid, '../topography/topo.sub150.tx2_3v3.SRTM15_V2.4.edit2.SmL1.0_C1.0.nc', min_depth=9.5, varname="D_edit2")
```

```python
# CICE grid file
fname_grid = "./cice_grid_tx2_3v3_{}.nc".format(today)
topo.write_cice_grid(fname_grid)
```

```python
fname_grid_cfd = "./cice_grid_tx2_3v3_{}_cdf5.nc".format(today)
```
